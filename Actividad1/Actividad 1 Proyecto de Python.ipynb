```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("diabetes.csv")

# Mostrar las primeras filas
print(df.head())
```

2. Limpieza y transformación de los datos

```python
# Información general
print(df.info())

# Reemplazar valores 0 en columnas que no pueden tener 0
cols_to_clean = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_to_clean] = df[cols_to_clean].replace(0, pd.NA)

# Rellenar valores faltantes con la mediana
for col in cols_to_clean:
    df[col] = df[col].fillna(df[col].median())

# Verificar que no haya valores nulos
print(df.isnull().sum())
```

3. Análisis exploratorio de datos (EDA)

```python
# Estadísticas generales
print(df.describe())

# Distribución de la variable objetivo
print(df['Outcome'].value_counts())
```

4. Visualización de resultados con pandas y matplotlib

Histogramas de las variables

```python
df.hist(figsize=(12, 10), bins=20)
plt.suptitle("Distribución de variables", fontsize=16)
plt.tight_layout()
plt.show()
```

 Cada histograma muestra la distribución de frecuencia de los valores de la variable

Gráfico de barras para la variable Outcome

```python
df['Outcome'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title("Distribución de diagnósticos (0=No, 1=Sí)")
plt.xlabel("Outcome")
plt.ylabel("Frecuencia")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Gráfico de dispersión: Glucosa vs BMI

```python
df.plot(kind='scatter', x='Glucose', y='BMI', alpha=0.5)
plt.title('Relación entre Glucosa y BMI')
plt.grid(True)
plt.show()
```
